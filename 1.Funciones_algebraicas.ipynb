```python
import numpy as np
import matplotlib.pyplot as plt
```

#### **Función lineal**
Tiene la forma de $$ f(x)=mx+b $$ 
donde $m$ y $b \in \mathbb{R}$ 
$m$ puede ser calculada por $$ m=\frac{y_2-y_1}{x_2-x_1}  $$
y $b$ es el punto de corte de corte con el eje $y$.

Su dominio es $$Dom_f=(-\infty, +\infty)$$


```python
def f(m,b,x):
    return m*x + b
m, b = 2,3

x = np.linspace(-10,10,num=1000)
y = f(m,b,x)
# plt.grid()
fig, ax = plt.subplots()
ax.plot(x, y, color='g')
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
plt.show()
```

#### **Funciones polinómicas**
Tiene la forma de: $$ P(x) = a_nx^n + a_{n-1}x^{x-1}+ ... + a_2x^2+a_1x + a_0 $$
**Por ejemplo:**
$$P(x)=2x^7-x^4+3x^2+4$$

```python
def f(x):
    return 1.5*x**3 + 3*x**2 + 3*x - 5
x = np.linspace(-10,10, num=100)
y = f(x)
fig, ax = plt.subplots()
plt.xlim(-2, 2)
plt.ylim(-f(2), f(2))
ax.plot(x, y, color='g')
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='red', linewidth='1')
ax.axvline(x=0, color='red', linewidth='1')
plt.show()
```
