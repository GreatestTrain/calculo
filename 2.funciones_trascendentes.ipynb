```python
import numpy as np
import matplotlib.pyplot as plt
```

Son funciones que no pueden ser expresadas con polinomios.

#### **Funciones trigonométricas**
Son funciones que utilizan relaciones entre los lados de un triángulo rectángulo.

```python
def f(x):
    return np.cos(x)

x = np.linspace(-10,10, num=500)
y = f(x)
fig, ax = plt.subplots()
ax.plot(x, y, color='g')
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
plt.show()
```

##### **Función exponencial**
Tienen la forma de:
$$ f(x) = a^x $$
donde la base $a$ es una constante positiva.
$$ f(x) = e^x $$

```python
def f(x):
    return np.exp(x)

x = np.linspace(-10,10,num=1000)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, color='g')
plt.xlim(-1,1)
plt.ylim(-1,5)
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
plt.show()
```

#### **Función logaritmo**
Tienen la forma de:
$$ f(x)=log_{b}(x) \Longleftrightarrow x = b^{f(x)}$$

donde:
+ $b$ es la base
+ $f(x)$ es el exponente
+ $x$ es el resultado de elevar v al exponente $f(x)$
**Ejemplo:**
Teninedo b=" y n= 8, entonces:
$$2^8=256$$
Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:
$$log_{2}(256) = 8$$

```python
def f(x):
    return np.log2(x)

x = np.linspace(0.01,10,num=1000)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, color='g')
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
plt.show()
```

#### **Función Seccionada**
Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside:
$$ 
f(x) = 
\begin{cases}
0,&\quad\text{para, }x<0\\
1,&\quad\text{para, }x\ge0\\
\end{cases}
$$

```python
def f(x):
    aux = np.zeros_like(x)
    for key, value in enumerate(x):
        if value < 0:
            aux[key] = 0
        elif value >= 0:
            aux[key] = 1
    return aux
    
x = np.linspace(-10,10,num=500)
y = f(x)
fig, ax = plt.subplots()
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1', linestyle=':')
ax.plot(x, y, color='g')
plt.show()
```
