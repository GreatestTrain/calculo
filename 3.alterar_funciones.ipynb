```python
import numpy as np
import matplotlib.pyplot as plt
```

#### **Desplazamientos**
Siendo $c$ una constante mayor que cero entonces:
+ $y=f(x)+c$ : Desplazamiento vertical hacia arriba.
+ $y=f(x)-c$ : Desplazamiento vertical hacia abajo.
+ $y=f(x+c)$ : Desplazamiento horizontal hacia la izquierda.
+ $y=f(x-c)$ : Desplazamiento horizontal hacia la derecha.

```python
def f(x):
    return x**2
c = 10
x = np.linspace(-25,25,num=100)
y = f(x+c)
fig, ax = plt.subplots()
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
ax.plot(x, y, color='g')
plt.show()
```

#### **Deformaciones**
Siendo $c$ una constante diferente de cero entonces:
+ $y=f(x)·c$ : Alargamiento Vertical
+ $y=f(x)·\frac{1}{c}$ : Compresión Vertical
+ $y=f(x·c)$ : Compresión Vertical
+ $y=f(x·\frac{1}{c})$ : Alargamiento Horizontal

```python
def f(x):
    return x**3
c = 2
x = np.linspace(-10,10,num=1000)
y = f(x/c)
fig, ax = plt.subplots()
plt.xlim(-2.5,2.5)
plt.ylim(-2.5,2.5)
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
ax.plot(x, y, color='g')
plt.show()
```
