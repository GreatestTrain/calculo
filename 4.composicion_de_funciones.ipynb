```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return x**2

def f(x):
    return np.sin(x)

x = np.linspace(-10,10, num=1000)
y = f(g(x))
fig, ax = plt.subplots()
ax.grid(color='r', linestyle='dotted', linewidth='1')
ax.axhline(y=0, color='black', linewidth='1')
ax.axvline(x=0, color='black', linewidth='1')
ax.plot(x, y, color='g')
plt.show()
```
