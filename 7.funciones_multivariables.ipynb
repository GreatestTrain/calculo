**Funciones multivariables**
-
---

Tienen la forma:
$$ y = f(x_1, x_2, ..., x_n) $$
Donde $y$ es la variable dependiente.

Solo presentan un resultado

#### **Ejemplo**
$$ z = f(x, y)$$

El dominio de una función es una región área formada por las variables independientes.

```python
## Importar librerias
from matplotlib import cm # manejo de colores y gradientes
import numpy as np
import matplotlib.pyplot as plt
```

#### **Gráfica de 3 dimensiones**

Función a graficar:
$$ f(x) = sin(x) + 2·cos(y) $$

```python
def f(x, y):
    return np.sin(x) + 2*np.cos(y)

### Variable: Resolucion
res = 1000
### Variables: x, y, z
x = np.linspace(-4,4,num=res)
y = np.linspace(-4,4,num=res)
### Variables: XX, YY -> Generar una matriz coordinada entre x y y
XX, YY = np.meshgrid(x,y)
### Variable: z -> Aplicando la funcion en z (usar los valores asignados por el método np.meshgrid)
z = f(XX,YY)
### Gráfico
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
surf = ax.plot_surface(XX,YY,z, cmap=cm.inferno)
fig.colorbar(surf)
plt.rcParams['figure.figsize']=10,6
plt.rcParams['figure.dpi']=72
plt.show()
```

#### **Curvas de nivel**
